```python
import statsmodels.api as sm
import pandas as pd

mtcars = sm.datasets.get_rdataset('mtcars').data
mtcars.head()
```

```python
# make a regression model to classify cars given mpg and am into two categories.
# here consider the variable vs as a response variable
# GLM (General Linear Model)
model = sm.GLM(mtcars['vs'], sm.add_constant(mtcars[['mpg','am']]), family=sm.families.Binomial()).fit()
print(model.summary())
```

Given that the coefficient of mpg is 0.6809, when mpg increases by one unit, the odds of vs=1 increases by exp(0.6809)&#126;1.968. 
<br>From the same reasoning, consering that the coefficient of am is -3.0073, the odd of vs=1 does by exp(-3.0073)&#126;0.05. Namely, as for am=0, the odds of vs=1 decreases by 95%, compared to am=1.
