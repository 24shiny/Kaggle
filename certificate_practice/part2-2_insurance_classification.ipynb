```python
import pandas as pd
train = pd.read_csv('data/Insurance_train_10.csv')
test_input = pd.read_csv('data/Insurance_test_10.csv')
train.head()
```

```python
test_input.head()
```

```python
train.isna().sum()
```

```python
test_input.isna().sum()
```

```python
print(train['Ever_Married'].unique())
print(train['Graduated'].unique())
print(train['Profession'].unique())
```

```python
# as there are quite many rows, drop ones having missing values
train_input = train.dropna()
test_input = test_input.dropna()

train.isna().sum()
```

```python
train_target = train['Segmentation'].values
train_input = train_input.drop(columns=['Segmentation']).reset_index(drop=True)
```

```python
train_target
```

```python
train_input.head()
```

```python
# LabelEncoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(train_input['Gender'])
train_input['Gender'] = le.transform(train_input['Gender'])
test_input['Gender'] = le.transform(test_input['Gender'])

le = LabelEncoder()
le.fit(train_input['Ever_Married'])
train_input['Ever_Married'] = le.transform(train_input['Ever_Married'])
test_input['Ever_Married'] = le.transform(test_input['Ever_Married'])

le = LabelEncoder()
le.fit(train_input['Graduated'])
train_input['Graduated'] = le.transform(train_input['Graduated'])
test_input['Graduated'] = le.transform(test_input['Graduated'])

le = LabelEncoder()
le.fit(train_input['Profession'])
train_input['Profession'] = le.transform(train_input['Profession'])
test_input['Profession'] = le.transform(test_input['Profession'])

le = LabelEncoder()
le.fit(train_input['Spending_Score'])
train_input['Spending_Score'] = le.transform(train_input['Spending_Score'])
test_input['Spending_Score'] = le.transform(test_input['Spending_Score'])
```

```python
train_input.head()
```

```python
test_input.head()
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_input = pd.DataFrame(ss.transform(train_input), columns=train_input.columns)
test_input = pd.DataFrame(ss.transform(test_input), columns=train_input.columns)

train_input.head()
```

```python
print(train_input.shape, len(train_target))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train_input, train_target)
print(rf.score(train_input, train_target))
```

```python
train_predict = rf.predict(train_input)
test_predict = rf.predict(test_input)
```

```python
from sklearn.metrics import f1_score
f1_score(train_target, train_predict, average='macro')
```

```python
submit = pd.DataFrame(test_predict)
submit = submit.rename(columns={0:'Segmentation_pred'})
submit.head()
```

```python
submit.to_csv('seg_classification.csv', index=False)
```
