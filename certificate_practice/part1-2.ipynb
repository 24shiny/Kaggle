# D1 : airquality of Seoul

```python
import pandas as pd
air = pd.read_excel('data/period1.xlsx')
air.head()
```

```python
air.info()
```

# preprocessing : check and drop missing values

```python
air.isna().sum()
```

```python
# rename columns 
air = air.rename(columns = {
    '날짜' : 'date',
    '측정소명' : 'region',
    '미세먼지 PM10\n(㎍/m3)' : 'PM10',
    '초미세먼지\nPM2.5 (㎍/m3)' : 'PM2.5',
})

air1 = air[['date', 'region', 'PM10', 'PM2.5']]
air1.head()
```

```python
# how many regions?
air1['region'].value_counts()
```

```python
air1['date'].unique()
```

```python
# remove redundants rows where date = '전체' and region ='평균'
air1 = air1[~((air1['date']=='전체')|(air1['region']=='평균'))]
air1.head()
```

```python
# drop missing values
air2 = air1.dropna()
air2.head()
```

```python
air2['PM10'].max()
```

```python
# Q1 : date, region and PM10 when PM10 is the highest
air2[air2['PM10']==air2['PM10'].max()]
```

```python
# Q2 : average of PM10 in Seoul from Feb to March and print the top five regions showing the highest values
air2.groupby('region')['PM10'].mean().reset_index()
```

```python
air_groupped = air2.groupby('region')['PM10'].mean().reset_index()
air_groupped.sort_values(by='PM10', ascending=False).head(5) # or sort_mean.head(5)
```

```python
# Q3 : create a new feature to classify PM10 into four; good (0-30), normal (31-80), bad (81-150), and bad (151-)
air2.loc[:,'grade'] = pd.cut(air2['PM10'], bins=[-float('inf'), 31, 81, 151, float('inf')],
                      labels = ['good', 'normal', 'bad', 'worse'], right=False)
air2.head()
```

```python
air_grade = air2['grade'].value_counts()
air_grade = air_grade.to_frame().reset_index()
air_grade
```

```python
air_grade['percentage'] = air_grade['count'] / len(air2)
air_grade.head()
```

```python
# Q4 : the region showing the highest PM10 on a date when PM2.5 is minimal
air3 = air2[(air['PM2.5']==air['PM2.5'].min())]
air3.head()
```

```python
air3.sort_values(by='PM10', ascending=False).head(3)
```

# D2 : Subway

```python
import pandas as pd
subway = pd.read_csv('data/CARD_SUBWAY_MONTH_202203.csv', encoding='cp949')
subway.head()
```

```python
subway.info()
```

```python
# rename columns and drop the last column
subway = subway.rename(columns={
    '사용일자' : 'date',
    '노선명' : 'line',
    '역명' : 'station',
    '승차총승객수' : 'on_pass',
    '하차총승객수' : 'off_pass'
}).drop(columns=['등록일자'])

subway.head()
```

```python
# average of on_pass and off_pass
# agg() maps a column to a function in the dictionary form
subway.agg({'on_pass':'mean', 'off_pass':'mean'})
```

```python
# which line has the most passengers (on_pass)?
subway[subway['on_pass']==subway['on_pass'].max()]
```

```python
# alternatively,
subway.loc[subway['on_pass'].idxmax(), ['line', 'station', 'on_pass']]
```

```python
# the top three stations in terms of the average of total passengers
subway['total_pass'] = subway['on_pass'] + subway['off_pass']
subway_groupped = subway.groupby('station')['total_pass'].mean().reset_index()
subway_groupped
```

```python
subway_groupped.sort_values(by='total_pass', ascending=False).head(3)
```

```python
# alternatively,
subway_groupped = subway.groupby('station').agg({'total_pass':'mean'})
subway_groupped.sort_values(by='total_pass', ascending=False).head(3)
```

# D3 salary

```python
salary = pd.read_csv('data/koweps.csv')
salary.head()
```

```python
salary.columns
```

```python
# rename columns of interest
salary = salary.rename(columns={
    'h1601_4':'gender', 'h1601_5':'birth', 'h1601_6':'edu',
    'h16_reg5':'region', 'h1608_114':'r_salary', 'h1608_122':'t_salary'
})

# and single them out
salary = salary[['gender', 'birth', 'edu', 'region', 'r_salary', 't_salary']]
salary.head()
```

```python
salary.isna().sum()
```

```python
import numpy as np

# np.where(codition, TRUE, FALSE)
# replace zero with np.nan in columns r_salary and t_salary
salary['r_salary'] = np.where(salary['r_salary']==0, np.nan, salary['r_salary'])
salary['t_salary'] = np.where(salary['t_salary']==0, np.nan, salary['t_salary'])

salary['age'] = 2021 - salary['birth'] + 1
print(salary['age'].min(), salary['age'].max())
```

```python
salary['edu'].unique()
```

```python
salary.loc[:,'edu_grade'] = pd.cut(salary['edu'], 
                                   bins=[-float('inf'), 5,6,7, float('inf')],
                                  labels=['중학이하','고교','전문대','대학이상'],
                                  right=False)
salary.head()
```

```python
salary['edu_grade'].value_counts()
```

```python
# alternatively,
salary['edu_grade'] = '대학이상'
salary.loc[salary['edu'].isin([2,3,4]), 'edu_grade'] = '중학이하'
salary.loc[salary['edu']==5, 'edu_grade'] = '고교'
salary.loc[salary['edu']==6, 'edu_grade'] = '전문대'
```

```python
# 3-1 female and male regular employees getting the highest salary, respectively
# drop rows having NaN first

salary_filtered = salary.dropna(subset=['r_salary']).groupby('gender')
salary_filtered.head()
```

```python
salary_filtered['r_salary'].idxmax()
```

```python
salary.loc[salary_filtered['r_salary'].idxmax()]
```

```python
# 3-2 ages of the top 3 having the highest average of r_salary
salary.dropna(subset=['r_salary']).groupby('age')['r_salary'].mean().reset_index()
```

```python
salary_age = salary.dropna(subset=['r_salary']).groupby('age')['r_salary'].mean().reset_index()
salary_age.sort_values(by='r_salary', ascending=False).head(3)
```

```python
# 3-3 average salary of regular employees with respect to education level and gender
salary.dropna(subset=['r_salary']).groupby(['gender','edu_grade'])['r_salary'].mean().reset_index()
```

```python
salary_gender_edu = salary.dropna(subset=['r_salary']).groupby(['gender','edu_grade'])['r_salary'].mean().reset_index()
salary_gender_edu.sort_values(by='r_salary', ascending=False)
```
