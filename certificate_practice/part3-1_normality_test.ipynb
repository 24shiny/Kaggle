Q1. normality test; Shapiro-Wilk test
<p>
    <ol>
        <li>H0 : samples follow the normal distribution</li>
        <li>H1 : samples do not follow the normal distribution</li>
    </ol>
</p>

```python
import pandas as pd
mtcars = pd.read_csv('data/mtcars.csv')
mtcars.head()
```

```python
mtcars.isna().sum()
```

```python
mtcars = mtcars.rename(columns={'Unnamed: 0':'model'})
# as categorical variables must not be put to the stats.shapiro(),
mtcars = mtcars.drop(columns=['model']).reset_index(drop=True)
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.describe()
```

```python
# do the test w.r.t 'am'
mtcars['am'].unique()
```

```python
# extract rows whose 'am' are one
mtcars1 = mtcars[mtcars['am']==1].reset_index(drop=True)
mtcars1.head()
```

```python
# do the shapiro test
from scipy import stats
sp = stats.shapiro(mtcars1)
sp
```

As the pvalue is less than the general significant level (0.05), we reject the null hypothesis and accept the alternative hypothesis.
In conclustion, the variable 'am' does not follow the normal distribution.
