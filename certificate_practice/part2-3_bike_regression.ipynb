```python
import pandas as pd
bike = pd.read_csv('data/bike_train.csv')
bike.head()
```

```python
import datetime as dt
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike['year'] = bike['datetime'].dt.year
bike['month'] = bike['datetime'].dt.month
bike['day'] = bike['datetime'].dt.day
bike['hour'] = bike['datetime'].dt.hour
```

```python
# drop redundant columns
drop_cols = ['datetime', 'atemp', 'casual', 'registered']
bike = bike.drop(columns=drop_cols).reset_index(drop=True)
bike.head()
```

```python
bike.isna().sum()
```

```python
train_target = bike['count'].values
train_input = bike.drop(columns=['count']).reset_index(drop=True)

print(train_target)
train_input.head()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(train_input, train_target, random_state=42)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(train_input, train_target)
train_predict = rf.predict(train_input)
test_predict = rf.predict(test_input)

print(rf.score(train_input, train_target))
print(rf.score(test_input, test_target))
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(test_predict, test_target)
```

```python
submission = pd.DataFrame(test_predict)
submission = submission.rename(columns={0:'count_pred'})
submission.head()
```

```python
submission.to_csv('count_regression.csv', index=False)
```
