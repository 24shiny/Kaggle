```python
import pandas as pd
train = pd.read_csv('data/score_train.csv')
test = pd.read_csv('data/score_test.csv')
train.head()
```

```python
test.head()
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train_score = train.pop('Credit_Score')
```

```python
# one-hot encoding
print(train.shape, test.shape)
combined = pd.concat([train, test])
combined = pd.get_dummies(combined)
train = combined[:len(train)]
test = combined[-len(test):]
print(train.shape, test.shape)

from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target = train_test_split(train, train_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train_input, train_target)
print(rf.score(train_input, train_target), rf.score(val_input, val_target))
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.head()
```

```python
submit.to_csv('7_multiple_classification_credit.csv', index=False)
```
