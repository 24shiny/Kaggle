```python
import pandas as pd
train = pd.read_csv('data/flight_train.csv')
test = pd.read_csv('data/flight_test.csv')
train.head()
```

```python
test.head()
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train_target = train['price']
train_input = train.drop(columns=['flight','price'])
test_input = test.drop(columns=['flight'])
```

```python
train_input.head()
```

```python
test_input.head()
```

```python
train = pd.get_dummies(train_input)

from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target = train_test_split(train, train_target)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(train_input, train_target)
print(rf.score(train_input, train_target), rf.score(val_input, val_target))
```

```python
test_input = pd.get_dummies(test_input)
pred = rf.predict(test_input)
submit = pd.DataFrame({'pred':pred})
submit.head()
```

```python
submit.to_csv('8_regression_flight.csv',index=False)
```
