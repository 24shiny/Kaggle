```python
import pandas as pd
train = pd.read_csv('data/hr_train.csv')
test = pd.read_csv('data/hr_test.csv')
train.head()
```

```python
test.head()
```

```python
train.info()
```

```python
print(train['major_discipline'].unique())
print(train['experience'].unique())
print(train['company_size'].unique())
print(train['company_type'].unique())
print(train['last_new_job'].unique())
```

```python
train['major_discipline'] = train['major_discipline'].fillna('No Major')
test['major_discipline'] = train['major_discipline'].fillna('No Major')

import numpy as np
train['experience'] = train['experience'].replace(to_replace=[np.nan, '<1', '>20'], value = [0, 0, 21])
train['experience'] = train['experience'].astype(int)
test['experience'] = test['experience'].replace(to_replace =[np.nan, '<1', '>20'], value = [0, 0 ,21])
test['experience']= test['experience'].astype(int)

train['company_size'] = train['company_size'].replace(to_replace =[np.nan, '<10', '10/49', '50-99', '100-500', '500-999', '1000-4999', '5000-9999', '10000+'],
                                                      value =[0,1,2,3,4,5,6,7,8])
train['company_size']= train['company_size'].astype(int)
test['company_size'] = test['company_size'].replace(to_replace =[np.nan, '<10', '10/49', '50-99', '100-500', '500-999', '1000-4999', '5000-9999', '10000+'],
                                                      value =[0,1,2,3,4,5,6,7,8])
test['company_size']= test['company_size'].astype(int)

train['company_type'] = train['company_type'].fillna('Other')
test['company_type'] = test['company_type'].fillna('Other')

train['last_new_job'] = train['last_new_job'].replace(to_replace =[np.nan, 'never', '>4'], value =[0,0,5])
train['last_new_job']= train['last_new_job'].astype(int)
test['last_new_job'] = test['last_new_job'].replace(to_replace =[np.nan, 'never', '>4'], value =[0,0,5])
test['last_new_job']= test['last_new_job'].astype(int)
```

```python
print(train['gender'].unique())
```

```python
train['gender']= train['gender'].map({'Male':1, 'Female':2, np.nan:0, 'Other':0})
test['gender']= test['gender'].map({'Male':1, 'Female':2, np.nan:0, 'Other':0})
print(train['gender'].unique())
```

```python
# drop : enrollee_id, city
train_target = train['target']
train = train.drop(columns=['enrollee_id', 'city','target'])
test = test.drop(columns=['enrollee_id', 'city'])
train.head()
```

```python
print(train['enrolled_university'].unique())
print(test['enrolled_university'].unique())
print(train['education_level'].unique())
print(test['education_level'].unique())
```

```python
train['enrolled_university']= train['enrolled_university'].fillna('no_enrollment')
train['enrolled_university']= train['enrolled_university'].replace([np.nan], ['no_enrollment'])
test['enrolled_university']= test['enrolled_university'].fillna('no_enrollment')
test['enrolled_university']= test['enrolled_university'].replace([np.nan], ['no_enrollment'])
train['enrolled_university']= train['enrolled_university'].map({'no_enrollment':0, 'Part time course':1, 'Full time course':2})
test['enrolled_university']= test['enrolled_university'].map({'no_enrollment':0, 'Part time course':1, 'Full time course':2})
# train['enrolled_university']= train['enrolled_university'].astype(int)
# test['enrolled_university']= test['enrolled_university'].astype(int)

train['education_level'] = train['education_level'].fillna('Graduate') # on average,
test['education_level'] = test['education_level'].fillna('Graduate')
train['education_level']= train['education_level'].replace([np.nan], ['Graduate'])
test['education_level']= test['education_level'].replace([np.nan], ['Graduate'])
train['education_level'] = train['education_level'].map({'Primary School':0, 'High School':1, 'Graduate':2, 'Masters':3, 'Phd':4})
test['education_level'] = test['education_level'].map({'Primary School':0, 'High School':1, 'Graduate':2, 'Masters':3, 'Phd':4})
# train['education_level']= train['education_level'].astype(int)
# test['education_level']= test['education_level'].astype(int)
```

```python
print(train['enrolled_university'].unique())
print(test['enrolled_university'].unique())
print(train['education_level'].unique())
print(test['education_level'].unique())
```

```python
train['enrolled_university']= train['enrolled_university'].astype(int)
test['enrolled_university']= test['enrolled_university'].astype(int)
train['education_level']= train['education_level'].astype(int)
test['education_level']= test['education_level'].astype(int)
```

```python
train.head()
```

```python
print(train.shape, test.shape)

combined = pd.concat([train, test])
combined = pd.get_dummies(combined)

train = combined[:len(train)]
test = combined[-len(test):]

print(train.shape, test.shape)
```

```python
from sklearn.model_selection import train_test_split
train_input, val_input, train_target_2, val_target = train_test_split(train, train_target)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train_input, train_target_2)
print(rf.score(train_input, train_target_2), rf.score(val_input, val_target))
```

```python
pred = rf.predict_proba(test)
submit = pd.DataFrame({'pred':pred[:,1]})
submit.head()
```

```python
submit.to_csv('6_binary_regression_turnover.csv', index=False)
```
